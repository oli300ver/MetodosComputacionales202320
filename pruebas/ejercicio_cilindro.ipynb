```python
import numpy as np
import matplotlib.pyplot as plt

# a) Definir una discretización en los ejes x e y
N = 25
x = np.linspace(-4, 4, N)
y = np.linspace(-4, 4, N)
X, Y = np.meshgrid(x, y)

# b) Definir la función potencial del flujo
V = 2  # cm/s
R = 2  # cm

def phi(x, y):
    return V * x* (1-(R**2/ np.sqrt(x**2 + y**2)))

# c) Calcular el campo de velocidades
h = 0.001
vx = np.gradient(phi(X, Y), h, axis=0)
vy = -np.gradient(phi(X, Y), h, axis=1)

# Establecer el campo de velocidades igual a cero dentro del cilindro (r < R)
inside_cylinder = X**2 + Y**2 < R**2
vx[inside_cylinder] = 0
vy[inside_cylinder] = 0

# d) Dibujar el campo de velocidades
fig, ax = plt.subplots(figsize=(8, 8))
ax.quiver(vx, vy, X, Y, scale=20, color='blue')

# Configuración de ejes y título
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Campo de Velocidades')

# Mostrar el gráfico
plt.show()


```
