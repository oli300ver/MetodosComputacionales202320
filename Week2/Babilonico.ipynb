```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 2.
n = 5
type(x)
```

```python
func = np.zeros(n)
type(func)
```

```python
func
```

```python
func[0] = x
```

```python
def Update(f,x):
    
    f = (x/f + f)/2.
    
    return f
    
```

```python
Update(x,x)
```

```python
for i in range(1,n):
    func[i] = Update(func[i-1],x)
```

```python
func
```

```python
k = np.arange(0,n,1)
```

```python
plt.scatter(k,func)
plt.axhline(y = np.sqrt(2),color='r')
plt.yscale('log')
```

```python
def GetDistance(x):
    return np.sqrt(x**2 + (3*x + 5)**2)
```

```python
x = np.linspace(-3,0,100)
y = GetDistance(x)
```

```python
plt.scatter(x,y)
```

```python
ii = np.where( y == np.amin(y) )
x[ii]
```

```python
def DerivadaCentral(f,x,h):
    
    d = 0.
    
    if h != 0:
        d = (f(x+h) - f(x-h))/(2*h)
        
    return d
```

```python
Derivada = DerivadaCentral(GetDistance,x,x[1]-x[0])
```

```python
plt.plot(x,Derivada)
```
